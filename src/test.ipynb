## Load front encoder

```python
import torch
from espnet.nets.pytorch_backend.transformer.encoder import Encoder
```

```python
path = "/home/zzzzz/VSR_LLM_AVSR/ckpts/auto_avsr/vsr_trlrwlrs2lrs3vox2avsp_base.pth"
```

```python
state_dict = torch.load(path, map_location="cpu")
print(state_dict.keys())
print(state_dict["encoder.frontend.trunk.layer1.0.conv1.weight"])
```

```python
encoder = Encoder(
            attention_dim=768,
            attention_heads=12,
            linear_units=3072,
            num_blocks=12,
            input_layer="conv3d",
            dropout_rate=0.1,
            positional_dropout_rate=0.1,
            attention_dropout_rate=0.1,
            encoder_attn_layer_type="rel_mha",
            macaron_style=True,
            use_cnn_module=True,
            cnn_module_kernel=31,
            zero_triu=False,
            a_upsample_ratio=1,
            relu_type="swish",
        )
```

```python
print(encoder.frontend.trunk.layer1[0].conv1.weight)
```

```python
print(encoder.state_dict().keys())
```

```python
for name, param in state_dict.items():
    try :
        if name.split("encoder.")[1] in encoder.state_dict():
            encoder.state_dict()[name.split("encoder.")[1]].copy_(param)
            print(f"the frontend param {name} is loaded")
    except Exception as e:
        print(e)
        print(f"the frontend param {name} is not loaded")
```
